```python
from google.colab import drive
import pandas as pd

# Mount Google Drive
drive.mount('/content/drive')

# Create an empty list to store the data
data = []

# Loop through the divs and extract the data
for i in range(1,60):
    # Check if anchor tag exists
    if divs[i].a:
        # Get the product link from the div
        product_link = divs[i].a['href']
    else:
        print("No anchor tag found in div index:", i)
        continue  # Skip to next iteration

    # Add the base URL if the href is relative
    if product_link.startswith('/'):
        product_link = 'https://www.cardekho.com' + product_link

    # Set headers to mimic a real browser
    headers = {
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }

    # Make the request with the headers
    response = requests.get(product_link, headers=headers)

    # Check the status code
    if response.status_code == 200:
        # Process the content as needed
        soup = bs(response.content, 'html.parser')
        html = soup.find_all('div', {'class': 'table-responsive'})
        if html:
            td_elements = html[0].tr.find_all('td')
            for td in td_elements:
                try:
                    title = td.div.find('span', class_='title').img['title']
                    price = td.find('div', class_='price').text.strip()
                    rating = td.find('div', class_="startRating")
                    if rating:
                        rating_star = rating.find('span', class_="ratingStarNew").text.strip()
                        review_text = rating.find('span', class_="reviewText").text.strip()
                    else:
                        rating_star = "null"
                        review_text = "null"
                    data.append({
                        "Title": title,
                        "Price": price,
                        "Rating": rating_star,
                        "Reviews": review_text
                    })
                except AttributeError:
                    data.append({
                        "Title": "null",
                        "Price": "null",
                        "Rating": "null",
                        "Reviews": "null"
                    })
    else:
        print(f"Failed to retrieve the page. Status code: {response.status_code}")

# Convert the data list to a pandas DataFrame
df = pd.DataFrame(data)

# Define the path where you want to save the file in Google Drive
file_path = '/content/drive/My Drive/New_car_data_1.csv'

# Save the DataFrame to a CSV file in Google Drive
df.to_csv(file_path, index=False)
```
